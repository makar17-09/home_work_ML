```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import random
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X = iris.data[50:]
y = iris.target[50:]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
def mse(actual, predicted):           #для вычисления ошибки
    return np.sum(np.square(predicted - actual)) / (2 * len(predicted))
```

```python
def accuracy_metric(actual, predicted):        #посчитать точность модели
    correct = 0
    for i in range(len(actual)):
        if actual[i] == predicted[i]:
            correct += 1
    return correct / float(len(actual)) * 100.0
```

```python
def logistic_regression(sample, params):        #модель логистической регрессии
    L = []
    for X in sample:
        L.append(params[0] + sum(params[i] * X[i - 1] for i in range(1, (len(X) + 1))))
    p = [1 / (1 + np.exp(-L_i)) for L_i in L]
    predictions = [1 if p_i < 0.5 else 2 for p_i in p]
    return predictions
```

```python
def gradient_procedure(lr, n_epochs, sample, y):       #метод ГС

    params = []
#     params = [0, 0, -1, 0.6, 0]
    costs = []
    params = np.random.normal(size=((len(sample[0]) + 1),))

    for _ in range(n_epochs):
        predictions = logistic_regression(sample, params)
        
        cost = mse(y, predictions)
        costs.append(cost)
        accuracy = accuracy_metric(y, predictions)
        
#         if accuracy > 85:
#             print(f"Epoch = {_}, Accuracy = {round(accuracy, 2)}, MSE = {round(cost, 2)}, Params: {params}")
#             return params, costs, _ + 1
        
        params[0] -= lr * np.sum(predictions - y) / len(predictions)
        for i in range(1, len(params)):
            params[i] -= lr * np.sum((predictions - y) * sample[:,i - 1]) / len(predictions)
        print(f"Epoch = {_}, Accuracy = {round(accuracy, 2)}, MSE = {round(cost, 2)}, Params: {params}")
    return params, costs, n_epochs
```

```python
def nesterov_momentum(lr, n_epochs, sample, y, nm):
    params = []
#     params = [0, 0, -1, 0.6, 0]
    costs = []
    params = np.random.normal(size=((len(sample[0]) + 1),))
    
    v_t = [0 for _ in range(len(params))]

    for _ in range(n_epochs):
        predictions = logistic_regression(sample, params)
        
        cost = mse(y, predictions)
        costs.append(cost)
        accuracy = accuracy_metric(y, predictions)
        
#         if accuracy > 85:
#             print(f"Epoch = {_}, Accuracy = {round(accuracy, 2)}, MSE = {round(cost, 2)}, Params: {params}")
#             return params, costs, _ + 1
        
        v_t[0] = nm * v_t[0] + lr * np.sum(predictions - y) / len(predictions)
        params[0] -= v_t[0]
        for i in range(1, len(params)):
            v_t[i] = nm * v_t[i] + lr * np.sum((predictions - y) * sample.T[i - 1]) / len(predictions)
            params[i] -= v_t[i]
        print(f"Epoch = {_}, Accuracy = {round(accuracy, 2)}, MSE = {round(cost, 2)}, Params: {params}")
    return params, costs, n_epochs
```

```python
def rmsprop(lr, n_epochs, sample, y, nm, e):
    params = []
#     params = [0, 0, -1, 0.6, 0]
    costs = []
    params = np.random.normal(size=((len(sample[0]) + 1),))
    
    Eg2 = [0 for _ in range(len(params))]
    
    for _ in range(n_epochs):
        predictions = logistic_regression(sample, params)
        
        cost = mse(y, predictions)
        costs.append(cost)
        accuracy = accuracy_metric(y, predictions)
        
#         if accuracy > 85:
#             print(f"Epoch = {_}, Accuracy = {round(accuracy, 2)}, MSE = {round(cost, 2)}, Params: {params}")
#             return params, costs, _ + 1
    
        Eg2[0] = nm * Eg2[0] + (1 - nm) * (np.sum(predictions - y) / len(predictions))**2
        params[0] -= (lr * np.sum(predictions - y) / len(predictions)) / np.sqrt(Eg2[0] + e)
        for i in range(1, len(params)):
            Eg2[i] = nm * Eg2[i] + (1 - nm) * (np.sum((predictions - y) * sample.T[i - 1]) / len(predictions))**2
            params[i] -= (lr * (np.sum((predictions - y) * sample.T[i - 1]) / len(predictions))) / np.sqrt(Eg2[i] + e)
        print(f"Epoch = {_}, Accuracy = {round(accuracy, 2)}, MSE = {round(cost, 2)}, Params: {params}")
    return params, costs, n_epochs
```

```python
n_epochs = 2001
lr = 0.01
nm = 0.9
e = 10e-6
```

```python
params_1, loss_array_1, epochs = gradient_procedure(lr, n_epochs, X_train, y_train)
```

```python
predictions_1 = logistic_regression(X_test, params_1)
cost_1 = mse(y_test, predictions_1)
accuracy_1 = accuracy_metric(y_test, predictions_1)
print(f"cost = {cost_1}, accuracy = {accuracy_1}")
```

```python
plt.plot(range(0, epochs), loss_array_1, 'r-')
plt.title('Ошибка')
plt.show()
```

```python
params_2, loss_array_2, epochs = nesterov_momentum(lr, n_epochs, X_train, y_train, nm)
```

```python
predictions_2 = logistic_regression(X_test, params_2)
cost_2 = mse(y_test, predictions_2)
accuracy_2 = accuracy_metric(y_test, predictions_2)
print(f"cost = {cost_2}, accuracy = {accuracy_2}")
```

```python
plt.plot(range(0, epochs), loss_array_2, 'r-')
plt.title('Ошибка')
plt.show()
```

```python
params_3, loss_array_3, epochs = rmsprop(lr, n_epochs, X_train, y_train, nm, e)
```

```python
predictions_3 = logistic_regression(X_test, params_3)
cost_3 = mse(y_test, predictions_3)
accuracy_3 = accuracy_metric(y_test, predictions_3)
print(f"cost = {cost_3}, accuracy = {accuracy_3}")
```
