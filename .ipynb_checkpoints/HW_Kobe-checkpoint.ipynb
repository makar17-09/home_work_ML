```python
def bar_plot(variable):

    var = data[variable]
    varValue = var.value_counts()

    plt.figure(figsize = (9,3))
    plt.bar(varValue.index, varValue)
    plt.xticks(varValue.index, varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
    print("{}: \n{}".format(variable,varValue))
```

```python
category1 = ["action_type","combined_shot_type","period","playoffs","season", "shot_made_flag", "game_date",
             "shot_type", "shot_zone_area", "shot_zone_basic", "shot_zone_range", "team_name", "matchup", "opponent"]
for c in category1:
    bar_plot(c)
```

Посмотрев на диаграммы можно заметить, что команда одна, значит можно удалить столбец team_name, а так же все столбы с id, "loc_x" и "loc_y" тоже можно удалить т.к. есть значение дистанции, дата не влияет на таргет. Т.к. shot_made_flag является целевой переменной и имеет много пропусков удалим строки с наличием пропуска

```python
data.drop(labels = ["game_event_id", "game_id", "team_id", "team_name", "game_date", 
                    "shot_id", "loc_x", "loc_y"], axis = 1, inplace = True)
```

```python
data = data.drop(data[data["shot_made_flag"].isnull()].index)
```

```python
data.info()
```

```python
import seaborn as sns
sns.pairplot(data)
```

```python
y = data['shot_made_flag']
```

```python
categorials_dummy = ["shot_type", "shot_zone_area", "shot_zone_basic", "shot_zone_range", "action_type","combined_shot_type", "season", "matchup", "opponent"]
dummy = pd.get_dummies(data[categorials_dummy])
```

```python
# from sklearn import preprocessing
# categorials_label = ["action_type","combined_shot_type", "season", "matchup", "opponent"]
# for cat in categorials_label:
#     labelencoder = preprocessing.LabelEncoder()
#     data[cat] = labelencoder.fit(data[cat]).transform(data[cat])
```

```python
categorials = ["lon", "minutes_remaining", "period", "playoffs", 
                "seconds_remaining", "shot_distance", "lat"]
```

```python
X = pd.concat([dummy, data[categorials]], axis=1)  #, data[categorials_label]
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.linear_model import LassoCV, RidgeCV
regrL = LassoCV()  #alpha=0.001
regrL.fit(X_train, y_train)
```

```python
regrL.score(X_train, y_train)
```

```python
regrL.score(X_test, y_test)
```

```python
ridge = RidgeCV()
```

```python
ridge.fit(X_train, y_train)
```

```python
ridge.score(X_train, y_train)
```

```python
ridge.score(X_test, y_test)
```

```python
dict_coef = {'Linear regression' : model.coef_, 'Lasso' : regrL.coef_, 'Ridge' : ridge.coef_}
```

```python
coef = pd.DataFrame(dict_coef)
```

```python
coef
```

```python
y = list(range(1, len(model.coef_) + 1))
```

```python
fig, ax = plt.subplots()
ax.plot(y, model.coef_, 'v', label = 'Linear regression')
ax.plot(y, regrL.coef_, 'x', label = 'Lasso')
ax.plot(y, ridge.coef_, 's', label = 'Ridge')
ax.legend()
fig.set_figheight(12)
fig.set_figwidth(15)
plt.show()
```

По графику заметно, что у линейной регрессии значения коэффицентов резко отличается от Лассо и Ридж. Коэффиценты Лассо не видно на графике, они почти занулены
