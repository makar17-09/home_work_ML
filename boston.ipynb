- CRIM уровень преступности на душу населения в разбивке по городам
- ZN доля жилой земли, зонированной для участков площадью более 25 000 кв. футов.
- INDUS Доля промышленных предприятий, не связанных с розничной торговлей, в акрах на город
- CHAS Фиктивная переменная реки ЧЕЗ Чарльз (= 1, если тракт ограничивает реку; 0 в противном случае)
- NOX Концентрация оксидов азота (частей на 10 миллионов)
- RM среднее количество комнат в одном жилом помещении
- AGE ВОЗРАСТНАЯ доля квартир, занятых владельцами, построенных до 1940 года
- DIS Взвешенные расстояния до пяти бостонских центров занятости
- RAD - индекс доступности радиальных магистралей
- TAX НАЛОГ на недвижимость по полной стоимости-ставка налога за 10 000 долларов США
- PTRATIO Соотношение учащихся и учителей в разбивке по городам
- B 1000(Bk - 0,63)^2, где Bk-доля чернокожих по городам
- LSTAT % более низкий статус населения
- MEDV Медианная стоимость домов, занятых владельцами, в МЕДВ в 1000 долларов США

С таргетной переменной коррелируют переменные: RM, ZN, B, DIS лучше всего. Попробуем обучить линейную модель на этих переменных

```python
sns.pairplot(data)
```

```python
select_col = ["ZN"]
dummy = pd.get_dummies(data[select_col], columns=select_col)
```

```python
from sklearn import preprocessing
categorials_label = ["CRIM", "B", "TAX", "LSTAT", "INDUS", "RM", "RAD"]
# categorials_label = ["CRIM", "B", "TAX", "ZN", "LSTAT", "INDUS", "RM"]
for cat in categorials_label:
    labelencoder = preprocessing.LabelEncoder()
    data[cat] = labelencoder.fit(data[cat]).transform(data[cat])
data["MEDV"] = labelencoder.fit(data["MEDV"]).transform(data["MEDV"])
```

```python
select_columns = ["DIS", "AGE", "PTRATIO", "CHAS", "NOX"]
# select_columns = ["DIS", "AGE", "PTRATIO", "CHAS", "NOX"]
y = data["MEDV"]
X = pd.concat([data[select_columns], data[categorials_label]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
